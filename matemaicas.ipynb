##Calculadora

```python
Cal = "bIENVENIDO A LA CALCULADORA:"
print(Cal)
num1 = int(input("elija su primer numero:"))
num2 = int(input("elija su seguno numero:"))
operacion = input("que operacion quieres escojer (+ - * /): ")

match operacion:
  case "+":
    resultado = num1+num2
  case "-":
    resultado = num1-num2
  case "*":
    resultado = num1*num2
  case "/":
    resultado = num1/num2
print(f"el resultado de {num1} {operacion} {num2} = {resultado} ")




```

##Tablas de multiplicar

```python
for i in range(1,20):
    print("esta es la tabla del ", i)
    for j in range(1,20):
        print(f"{i} x {j} = {i*j}")
    print("\n")
```
